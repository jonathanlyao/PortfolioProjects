```python
# import all necessary libraries such as numpy, pandas, matplotlib, seaborn, etc.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset into a pandas dataframe from the csv file using the read_csv() function
df = pd.read_csv('~\Documents\Projects\Most popular 1000 Youtube videos.csv')

# Display the first 5 rows of the dataframe using the head() function
df.head()
```

```python
# Display the last 5 rows of the dataframe using the tail() function
df.tail()
```

```python
# Check the data types of each column in the dataframe using the info() function
df.info()
```

```python
# Now we see that Video views, Likes and Dislikes are objects, so we need to convert them to floats
df['Video views'] = df['Video views'].str.replace(',', '').astype(float)
df['Likes'] = df['Likes'].str.replace(',', '').astype(float)
df['Dislikes'] = df['Dislikes'].str.replace(',', '').astype(float)
```

```python
# We see that 'Category' is an object, so we need to convert it to a category
df['Category'] = df['Category'].astype('category')
```

```python
# Check the data types of each column in the dataframe using the info() function again to make sure the conversion was successful
df.info()
```

```python
# Check for duplicate rows in the dataframe using the duplicated() function
df.duplicated().sum()
```

```python
# Check the number of rows and columns in the dataframe using the shape attribute
df.shape
```

```python
# Check for missing values in the dataframe using the isnull() function
df.isnull().sum()
```

```python
# Now we see that there are 473 missing values in the 'Dislikes' column, so we need to fill them with 0
df['Dislikes'] = df['Dislikes'].fillna(0)
```

```python
# There are 18 missing values in the 'Category' column, so we need to fill them with the most frequent category
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
```

```python
# Check for missing values in the dataframe again using the isnull() function just to make sure that there are no missing values
df.isnull().sum()
```

```python
# Check if there are any inconsistencies in the 'Category' column using the value_counts() function
df['Category'].value_counts()
```

```python
# Check the summary statistics of the numerical columns in the dataframe using the describe() function
df.describe()
```

```python
# Now check if any outliers are present in the numerical columns using box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Video views', 'Likes', 'Dislikes']])
plt.title('Box Plot of Video views, Likes and Dislikes')
plt.show()
```

```python
# It seems like there are outliers in the 'Video views', 'Likes' and 'Dislikes' columns, so we need to remove them
# We will remove the outliers using the IQR method
Q1 = df[['Video views', 'Likes', 'Dislikes']].quantile(0.25)
Q3 = df[['Video views', 'Likes', 'Dislikes']].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[['Video views', 'Likes', 'Dislikes']] < (Q1 - 1.5 * IQR)) | (df[['Video views', 'Likes', 'Dislikes']] > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# Check the outliers again using box plots
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Video views', 'Likes', 'Dislikes']])
plt.title('Box Plot of Video views, Likes and Dislikes')
plt.show()
```

```python
# Add a column 'Like/Dislike ratio' to the dataframe which is the ratio of Likes to Dislikes
df['Like/Dislike ratio'] = df['Likes'] / df['Dislikes']
```

```python
# Check the first 5 rows of the dataframe to see the new column
df.head()
```

```python
# Now we have all the data cleaned and ready for analysis
# We can now start analyzing the data to answer the questions we have
# Let's start by analyzing the what are the top 10 most viewed videos
top_10_most_viewed_videos = df.nlargest(10, 'Video views')
top_10_most_viewed_videos
```

```python
# What are the top 10 most liked videos
top_10_most_liked_videos = df.nlargest(10, 'Likes')
top_10_most_liked_videos
```

```python
# What video category has the highest average number of views and likes
category_avg_views_likes = df.groupby('Category')[['Video views', 'Likes']].mean()

# Now we can visualize the data using bar plots, by descending order of average number of views
plt.figure(figsize=(12, 6))
category_avg_views_likes['Video views'].sort_values(ascending=False).plot(kind='bar')
plt.title('Average Number of Views by Category')
plt.xlabel('Category')
plt.ylabel('Average Number of Views')
plt.show()
```

```python
# Check to see if there is a correlation between the number of views and likes, add a line to show the trend
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Video views', y='Likes', data=df)
plt.title('Correlation between Views and Likes')
plt.xlabel('Video views')
plt.ylabel('Likes')
plt.plot(np.unique(df['Video views']), np.poly1d(np.polyfit(df['Video views'], df['Likes'], 1))(np.unique(df['Video views'])), color='red')
plt.show()
```

```python
# Check to see how do engagaement metrics like Likes vary by category
plt.figure(figsize=(12, 6))
sns.boxplot(x='Category', y='Likes', data=df)
plt.title('Likes by Category')
plt.xlabel('Category')
plt.ylabel('Likes')
plt.xticks(rotation=90)
plt.show()
```

```python
# Check to see what's the average engagement rate (Likes/Views) by category
df['Engagement rate'] = df['Likes'] / df['Video views']
category_avg_engagement_rate = df.groupby('Category')['Engagement rate'].mean()
category_avg_engagement_rate
```

This prove that the music is the most popular category on Youtube
It has the highest average number of views and likes, and also the highest average engagement rate
This is consistent with the fact that music videos are one of the most popular types of content on Youtube

```python
# Check to see if the video title length has any impact on the number of views, add a line to show the trend
df['Title length'] = df['Video'].str.len()
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Title length', y='Video views', data=df)
plt.title('Impact of Title Length on Views')
plt.xlabel('Title length')
plt.ylabel('Video views')
plt.plot(np.unique(df['Title length']), np.poly1d(np.polyfit(df['Title length'], df['Video views'], 1))(np.unique(df['Title length'])), color='red')
plt.show()
```

Apparently, there is no impact between video title length and the number of views. 

```python
# Check to see the trend between the years and the number of views
yearly_views = df.groupby('published')['Video views'].sum()
plt.figure(figsize=(12, 6))
yearly_views.plot(kind='line')
plt.title('Trend of Views by Year')
plt.xlabel('Year')
plt.ylabel('Views')
plt.show()
```

```python
# # Check to see the trend between the years and the number of likes
yearly_like_dislike_ratio = df.groupby('published')['Likes'].sum()
plt.figure(figsize=(12, 6))
yearly_like_dislike_ratio.plot(kind='line')
plt.title('Trend of Likes by Year')
plt.xlabel('Year')
plt.ylabel('Likes')
plt.show()
```
